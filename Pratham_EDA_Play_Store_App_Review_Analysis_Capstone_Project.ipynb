```python
# Now we are filling missing value of rating with most average rating 
```

```python
play_store['Rating'].mode()
```

```python
play_store['Rating'].fillna(value=4.4, inplace=True)
```

```python
# Now filling the nan with Free app
```

```python
play_store['Type'].fillna(value='Free',inplace=True)
```

```python
# Now Installs column contain special char - '+'
#     Price column contain special char - '$'
#     It time to remove those special char
```

```python
# List of characters to remove
chars_to_remove = ['+',',','$']
# List of column names to clean
cols_to_clean = ['Installs','Price']

# Loop for each column in cols_to_clean
for col in cols_to_clean:
    # Loop for each char in chars_to_remove
    for char in chars_to_remove:
        # Replace the character with an empty string
        play_store[col] = play_store[col].apply(lambda x: x.replace(char,''))
```

```python
# Again Re moving Special char from size and converting kb into mb
# MB - M
# KB - k
```

```python
play_store['Size'] = play_store['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
play_store['Size'] = play_store['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
play_store['Size'] = play_store['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
play_store['Size'] = play_store['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
play_store['Size'] = play_store['Size'].apply(lambda x: float(x))
```

```python
# Now Converting Install, Price and Reviews datatype to int
```

```python
play_store['Installs'] = play_store['Installs'].apply(lambda x: int(x))
play_store['Price'] = play_store['Price'].apply(lambda x: float(x))
play_store['Reviews'] = play_store['Reviews'].apply(lambda x: int(x))
play_store.info()
```

```python
# Now we already deal with most of null values
# its time to deal with duplicates. we just gone drop duplicate by using drop_duplicate() function
```

```python
play_store.drop_duplicates(inplace=True)
```

# Top Category in terms of Number of Application

```python
category = play_store['Category'].value_counts().reset_index().rename(columns={'index':'Category','Category':'Count'})
```

```python
plt.figure(figsize= (14,7))
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.xlabel('Category')
plt.title("Number of Apps in each Category")
sns.barplot(category.Category,category.Count)
plt.show()
```

```python
# As we can see Family, Game, Tools has most number of applicaton avilable in store
```

# Top Category in terms of most number of install

```python
cat_installs = play_store.groupby('Category')['Installs'].sum().reset_index()
category_installs = cat_installs.sort_values(by=['Installs'],ascending=True)
```

```python
plt.figure(figsize= (14,7))
plt.xticks(rotation=90)
plt.xlabel('Installs')
plt.ylabel('Category')
plt.title("Number of Installs in each Category")
sns.barplot(category_installs.Category,category_installs.Installs)
plt.show()
```

```python
# This tell us the category of app that have most install is The Game,communciation,socail
```
